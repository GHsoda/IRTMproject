# process Json from Flair NER

```python
import json
data = []
with open('entities.json', 'r') as f:
    data = json.load(f)
```

```python
# remove 'CARDINAL', 'LANGUAGE', 'ORDINAL', 'QUANTITY', 'MONEY', 'TIME', 'PERCENT' tags in json
tags_to_remove = ['CARDINAL', 'LANGUAGE', 'ORDINAL', 'QUANTITY', 'MONEY', 'TIME', 'PERCENT']
filtered_data = [entity for entity in data if entity['value'] not in tags_to_remove]

with open('filtered_entities.json', 'w') as f:
    json.dump(filtered_data, f, indent=4)
```

```python
# check date entities with only year
import re

date_entities = [entity for entity in data if entity['value'] == 'DATE']

print(date_entities[:10])
print(len(date_entities))
```

```python
# create a year key in the date entities
decade_dict = {
    'twenties': 1920,
    'thirties': 1930,
    'forties': 1940,
    'fifties': 1950,
    'sixties': 1960,
    'seventies': 1970,
    'eighties': 1980,
    'nineties': 1990
}


for entity in date_entities:
    match = re.search(r'\b\d{4}\b', entity['entity'])
    if match:
        entity['year'] = int(match.group())
    else:
        for decade, year in decade_dict.items():
            if decade in entity['entity'] and not re.search(r'\bher\b', entity['entity']):
                if 'early' in entity['entity'] or 'beginning' in entity['entity']:
                    entity['year'] = year
                elif 'late' in entity['entity']:
                    entity['year'] = year + 9
                else:
                    entity['year'] = year + 5
```

```python
year_entities = [entity for entity in date_entities if 'year' in entity]
```

```python
# replace the Date entities
# import json

# with open('/content/drive/MyDrive/anna/filtered_entities.json', 'r') as f:
with open('filtered_entities.json', 'r') as f:
    data = json.load(f)
# exclude date entities and anna entities
non_date_anna_entities = [entity for entity in data if entity['value'] != 'DATE' and not entity['entity'].startswith('Anna')]

combined_data = non_date_anna_entities + year_entities


combined_data
```

```python
# standardize the Person entities
from difflib import SequenceMatcher
import copy

def similarity(a, b):
    return SequenceMatcher(None, a, b).ratio()

processed_data = copy.deepcopy(combined_data)
for entry in processed_data:
    if entry['value'] == 'PERSON' and entry['entity'].endswith('’s'):
        entry['entity'] = entry['entity'][:-2]

for i in range(len(processed_data)):
    if processed_data[i]['value'] == 'PERSON':
        for j in range(i+1, len(processed_data)):
            if processed_data[j]['value'] == 'PERSON':
                if abs(processed_data[i]['sentence_num'] - processed_data[j]['sentence_num']) <= 50:
                    if (processed_data[i]['entity'] in processed_data[j]['entity']) or (processed_data[j]['entity'] in processed_data[i]['entity']):
                        if len(processed_data[i]['entity']) > len(processed_data[j]['entity']):
                            processed_data[j]['entity'] = processed_data[i]['entity']
                        else:
                            processed_data[i]['entity'] = processed_data[j]['entity']
                    elif similarity(processed_data[i]['entity'], processed_data[j]['entity']) > 0.5:
                        if len(processed_data[i]['entity']) > len(processed_data[j]['entity']):
                            processed_data[j]['entity'] = processed_data[i]['entity']
                        else:
                            processed_data[i]['entity'] = processed_data[j]['entity']
```

```python
# correct 'Condé Nast' entities
for entry in processed_data:
    if entry['entity'] == 'Condé Nast':
        entry['value'] = 'ORG'
```

```python
# correct 'Harper’s Bazaar' entities
for entry in processed_data:
    if entry['entity'].endswith('Bazaar') or entry['entity'] == "Bazaar\u2019s":
        entry['entity'] = 'Harper\u2019s Bazaar'
        entry['value'] = 'ORG'
```

```python
# correct 'Novogrod' entities
for entry in processed_data:
    if entry['entity'] == 'Novogrod' and entry['value'] != 'PERSON':
        entry['entity'] = 'Nancy Novogrod'
        entry['value'] = 'PERSON'
```

```python
with open('streamgraph_entities.json', 'w') as f:
    json.dump(processed_data, f, indent=4)
```

```python
import pandas as pd
import numpy as np
import altair as alt

# Convert to DataFrame
df = pd.DataFrame(sorted_data)

# Forward fill NaNs in 'year' column
df['year'] = df['year'].fillna(method='ffill')

# Convert 'year' to integer
df['year'] = df['year'].astype('Int64')  # 'Int64' (capital "I") can handle NaN values

# Group by decade
df['decade'] = (df['year'] // 10 * 10)

# Filter out 'DATE' entities
df_entity = df[df['value'] != 'DATE']

# Create a DataFrame to include both 'value' and 'entity' fields
df_entity_value_counts = df_entity.groupby(['decade', 'value', 'entity']).size().reset_index(name='counts')

# Get the top 5 entities in each decade
top_entities_per_value = df_entity_value_counts.groupby(['decade', 'value']).apply(lambda x: x.nlargest(2, 'counts')).reset_index(drop=True)

# Create area chart
area_chart = alt.Chart(top_entities_per_value).mark_area().encode(
    alt.X('decade:O', 
          axis=alt.Axis(title='Decade', grid=True, gridColor='#ddd', labelAngle=0)), # Horizontal labels on the x-axis
    alt.Y('counts:Q', 
          stack='center', 
          axis=alt.Axis(title='Counts', labels=False, ticks=False, grid=False)), # Hide y-axis labels and ticks, remove gridlines
    alt.Color('value:N', 
              scale=alt.Scale(scheme='category20b'), 
              legend=alt.Legend(title="Entity Types")),
    tooltip=['entity', 'value', 'counts']  # Add tooltips
).properties(
    width='container',  # Adjust width to container size
    height=400  # Adjust height
).interactive()

area_chart.display()
```
