# Read the original text from the book

```python
# Read the original text from the book

# Open the text file in read mode
file = open('/content/drive/MyDrive/anna/origin.txt', 'r')

# Read the entire content of the file
content = file.read()
test_text = """
ORIGINS


Born Eleanor Baker to a wealthy Quaker family in 1917 in Harrisburg, Pennsylvania, the future Nonie Wintour was a society girl. Her father, Ralph Baker, was a lawyer who left private practice to become a Harvard Law School professor. He specialized in trusts, and, before his death, established a substantial fund that would support his descendants, including his granddaughter Anna, over the course of many decades.
"""
# Close the file
file.close()

# Print the content
print(content[:100])
print(len(content))
```

```python
# divide the book into chapters
import re

# use regular expression to find chapter headers
pattern = r'\n\n(Chapter \d+)'

# split the text into chapters by headers
sections = re.split(pattern, content)

# Remove leading/trailing whitespace and group chapters with their content
chapters = [sections[i] + sections[i + 1] for i in range(0, len(sections) - 1, 2)]
print(chapters[1][:100])
```

# get processed entity list

```python
# import 
import json
sorted_data = []
with open('/content/drive/MyDrive/anna/sorted_NER.json', 'r') as f:
    sorted_data = json.load(f)
f.close()
```

```python
# Create an empty list for each sentence in your text
num_sentences = max(item['sentence_num'] for item in sorted_data) + 1
named_entities = [[] for _ in range(num_sentences)]

# Add the named entities to the corresponding sentence's list
for item in sorted_data:
    named_entities[item['sentence_num']].append(item['entity'])
named_entities = [entities for entities in named_entities if entities]

print(named_entities)
```


# use Bert for preprocessing

```python
from sentence_transformers import SentenceTransformer

# Initialize the BERT model (note: this could take a while as it downloads the model)
sbert_model = SentenceTransformer('bert-base-nli-mean-tokens')

# Create document embeddings
doc_embeddings = sbert_model.encode(chapters)
```

```python
from bertopic import BERTopic

# Initialize BERTopic
topic_model = BERTopic(nr_topics=20, 
                       min_topic_size=10, 
                       calculate_probabilities=True, verbose=True)

# Fit model to the data
topics, probs = topic_model.fit_transform(chapters)

# Get the most frequent topics
topic_freq = topic_model.get_topic_info(); topic_freq.head(5)
```

# use LDA

```python
# preprocessing chapters and remove stopwords
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
import string
nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))


chapters_tokenized = []
for chapter in chapters:
    word_tokens = word_tokenize(chapter)
    filtered_sentence = [w for w in word_tokens if not w in stop_words and w not in string.punctuation]
    chapters_tokenized.append(filtered_sentence)

# Now 'chapters_tokenized' is a list of chapters with stop words removed
```

```python
chapters_tokenized[:3]
```

```python
# !pip install gensim
from gensim.corpora import Dictionary
from gensim.models import LdaModel

# Assuming named_entities is a list of lists, 
# where each inner list represents a document and contains the named entities detected in that document

dictionary = Dictionary(named_entities)
corpus = [dictionary.doc2bow(doc) for doc in named_entities]

lda_model = LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=16,  
    random_state=100,
    passes=20
)


# Print the topics
topics = lda_model.print_topics(num_words=5)
for topic in topics:
    print(topic)
```

```python
# !pip install pyLDAvis
import pyLDAvis
from pyLDAvis import gensim_models

pyLDAvis.enable_notebook()
visualization = gensim_models.prepare(
    lda_model, 
    corpus,
    dictionary, 
    mds = "mmds", 
    R = 30)

visualization
```
